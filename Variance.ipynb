# Дисперсия
Определение из Википедии:
Дисперсия — мера разброса случайной величины. Дисперсия случайной величины X обозначается как Var(X).
Пусть у нас есть случайная величина (Random variable) X. 

```python
import statistics
# Imports
from matplotlib.pyplot import hist
from pandas import DataFrame, Series
from seaborn import displot
```

```python
# Exploring data
x = range(-10, 11)
data: DataFrame = DataFrame({'X': x})
data: DataFrame = data.assign(Y=lambda df: df['X'] ** 2)
data['Const']: Series = 1
data
```

```python
bins: int = max(data['X']) - min(data['X']) + 1
values, bins, bars = hist(x = data['X'], bins=bins)
pass
```

```python
x = displot(data=data['X'], bins=bins, kind='hist')
```

```python
bins: int = max(data['Y']) - min(data['Y']) + 1
hist(x = data['Y'], bins=bins)
```

```python
displot(data=data['Y'], bins=bins, kind='hist')
```

```python
statistics.variance(data=data['X'])
```

```python
statistics.pvariance(data=data['X'])
```

```python
statistics.variance(data=data['Y'])
```

```python
statistics.variance(data=data['Const'])
```
